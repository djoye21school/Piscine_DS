##  make the aggregations with the car number

```python
top_count = df.groupby(['CarNumber'])['Fines']\
        .count() \
        .reset_index() \
        .sort_values('Fines', ascending=False)\
        .set_index('CarNumber')
```

```python
top_count
```

```python
df[df.index == top_count.head(1).index[0]]
```

```python
top_sum = df.groupby(['CarNumber'])['Fines']\
        .sum() \
        .reset_index() \
        .sort_values('Fines', ascending=False)\
        .set_index('CarNumber')
```

```python
top_sum
```

```python
df[df.index == top_sum.head(1).index[0]]
```

```python
df.groupby(['CarNumber', 'Make', 'Model']).count()
```

```python
df_dup = df[['Make', 'Model']].reset_index()
```

```python
df_dup.drop_duplicates(inplace=True)
```

```python
dup = df_dup.groupby(['CarNumber'])['Make']\
        .count()\
        .reset_index()\
        .sort_values('Make', ascending =  False)
```

```python
df[df.index == dup[dup.Make > 1]['CarNumber'].values[0]]
```
